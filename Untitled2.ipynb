### imports

```python
import re
import random
import pandas as pd
from nltk.stem import *
from sklearn import metrics
from nltk.corpus import stopwords
from spellchecker import SpellChecker
```

### global variables and objects

```python
words=stopwords.words('english')
words.extend(["m","im","every","so","father","mother","parent","son","daughter","child","husband",
             "wife","spouse","brother","sister","sibling","grandfather","grandmother","grandparents",
             "grandson","granddaughter","grandchild","uncle","aunt","nephew","niece","cousin"])
negprob={}
posprob={}
count={}
poscount={}
negcount={}
stemmer = SnowballStemmer("english")
spell=SpellChecker()
wrongwords={}
msg=[]
```

### All functions

```python
def train_test_split_list(value,label,test_size=0.3,seed=0):
    xtrain=[]
    xtest=[]
    ytrain=[]
    ytest=[]
    i=int(test_size*len(label))
    j=0
    while j<i:
        l=len(label)
        random.seed(seed)
        index=random.randint(0,l-1)
        xtest.append(value.pop(index))
        ytest.append(label.pop(index))
        j+=1
    xtrain=value
    ytrain=label
    return xtrain,xtest,ytrain,ytest

def read(filename,encoding="utf8"):
    global msg
    file=open(filename,'r',encoding=encoding)
    msg= file.readlines()
    x=[]
    y=[]
    for i in msg:
        try:
            l,v=i.split(',')
            x.append(v)
            y.append(int(l))
        except:
            y.append(0)
    return x,y

def extract_emo(line):
    y=re.sub(r'(.)\1{2,}',r'\1',line)
    y=re.sub('\s+',"",y)
    y=re.sub('\.+',"",y)
    y=re.findall(r'\W{2,}',y)
    return y

def remove_link(line):
    x=re.sub(r'http:.*$','',line)
    return x.strip()

def remove_time_stamp(line):
    x=re.sub(r'\d+:\d+','',line)
    return x.strip()

def remove_tag(line):
        y=re.sub(r'@\w+','',line)
        y=str.lower(y)
        return y.strip()

def remove_hashtag(line):
    y=re.sub(r'#\w+','',line)
    y=str.lower(y)
    return y.strip()

def remove_symbols(line,replacement):
    y=re.sub(r'\W+',replacement,line)
    y=re.sub(r'\s{2,}',"\s",y)
    return y.strip()

def remove_numbers(line):
    y=re.sub(r'3','e',line)
    y=re.sub(r'5','s',line)
    y=re.sub(r'\d+',' ',line)
    return y.strip()

def shorten(line):
    y=re.sub(r'(.)\1{2,}',r'\1',line)
    return y.strip()

def remove_stopwords(line):
    global words
    y=""
    for x in line.split():
        if x not in words:
            y+=x+" "
    return y.strip()

def preprocess(line):
    global stemmer,spell,wrongwords
    line=str.lower(line)
    line=remove_stopwords(line)
    line=remove_link(line)
    line=remove_time_stamp(line)
    line=remove_tag(line)
    line=remove_hashtag(line)
    line=remove_numbers(line)
    line=remove_symbols(line,' ')
    line=shorten(line)
    line=remove_stopwords(line)
    y=line.split()
    misspelled=spell.unknown(y)
    for i in misspelled:
        y.remove(i)
        if i not in wrongwords:
            wrongwords[i]=spell.correction(i)
        y.append(wrongwords[i])
    y=[stemmer.stem(j) for j in y]
    line=""
    for word in y:
        line+=word+" "
    return line.strip()

def train(xtrain,ytrain,ttype="text"):
    global count,negprob,posprob,poscount,negcount,words    
    for index,line in enumerate(xtrain):
        if ttype=="text":
            line=preprocess(line)
        y=line.split()                                                 
        for j in y:
            if j not in count:
                count[j]=1
            else:
                count[j]+=1
            if ytrain[index]==1:            
                if j not in poscount:
                    poscount[j]=1
                else:
                    poscount[j]+=1
            elif ytrain[index]==0:
                if j not in negcount:
                    negcount[j]=1
                else:
                    negcount[j]+=1
    for i in count:
        if i in poscount:
            posprob[i]=poscount[i]/count[i]
            negprob[i]=1-posprob[i]
        elif i in negcount:
            negprob[i]=negcount[i]/count[i]
            posprob[i]=1-negprob[i]
            

def test(xtest):
    global count,negprob,posprob
    ypred=[]
    for line in xtest:
        y=extract_emo(line)
        line=preprocess(line)
        y.extend(line.split())
        pos_pred_prob={}
        neg_pred_prob={}
        pcount=0
        ncount=0
        neutcount=0
        for i in y:
            if i not in posprob:
                pos_pred_prob[i]=0
            if i not in negprob:
                neg_pred_prob[i]=0
            if i in posprob:
                pos_pred_prob[i]=posprob[i]
                neg_pred_prob[i]=1-posprob[i]
            if pos_pred_prob[i]>neg_pred_prob[i]:
                pcount+=1
            elif pos_pred_prob[i]<neg_pred_prob[i]:
                ncount+=1
            else:
                neutcount+=1
        if pcount+neutcount>ncount:
            ypred.append(1)
        elif ncount>pcount and ncount>neutcount:
            ypred.append(0)
        else:
            pos=0
            neg=0
            for i in pos_pred_prob:
                pos+=pos_pred_prob[i]
                neg+=neg_pred_prob[i]
            if pos>neg:
                ypred.append(1)
            elif pos < neg:
                ypred.append(0)
            else:
                ypred.append(1)
    return ypred
```

### Reading the dataset and splitting to test and train set

```python
for x in words:
    words.remove(x)
    words.append(remove_symbols(x,""))
emx,emy=read('all_emo.csv')
train(emx,emy,ttype="emoji")
x,y=read('train.csv',encoding="latin1")
xtrain,xtest,ytrain,ytest=train_test_split_list(x,y,test_size=0.3,seed=10)
```

### Simple probablistic approach

```python
train(xtrain,ytrain)
pred=test(xtest)
oacc=metrics.accuracy_score(ytest,pred)
oprec=metrics.precision_score(ytest,pred)
orec=metrics.recall_score(ytest,pred)
print("accuracy=",oacc)
print("Recall=",orec)
print("Precision=",oprec)
print("Confusion Matrix\n",metrics.confusion_matrix(ytest,pred))
```

### Reading data set for MNB,SVC and Random forest and splitting them to test and train set

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
import pandas as pd

msg=pd.read_csv("train.csv",names=["label","value"],encoding="latin1")
emoji=pd.read_csv("all_emo.csv",names=["label","value"])
for i,x in msg["value"].items():
    c=extract_emo(x)
    m=preprocess(x)
    for emos in c:     
        m+=" "+emos
    msg.at[i,"value"]=m

xtrain1,xtest1,ytrain1,ytest1=train_test_split(msg.value,msg.label,test_size=0.2)

cv= CountVectorizer()
xtr_dtm=cv.fit_transform(xtrain1.values.astype('U'))
xte_dtm=cv.transform(xtest1.values.astype('U'))
xemo=cv.fit_transform(emoji.value)

```

### Multinomial Naive Bayes Algorithm

```python
from sklearn.naive_bayes import MultinomialNB

clf=MultinomialNB()
clf.fit(xemo,emoji.label)
clf.fit(xtr_dtm,ytrain1)
MNBpred= clf.predict(xte_dtm)

MNBacc=metrics.accuracy_score(ytest1,MNBpred)
MNBrec=metrics.recall_score(ytest1,MNBpred)
MNBprec=metrics.precision_score(ytest1,MNBpred)
print("accuracy=",MNBacc)
print("Recall=",MNBrec)
print("Precision=",MNBprec)
print("Confusion Matrix\n",metrics.confusion_matrix(ytest1,MNBpred))
```

### Support vector classifier

```python
from sklearn.svm import SVC
svclassifier=SVC(kernel='linear')
svclassifier.fit(xtr_dtm,ytrain1)
ypred=svclassifier.predict(xte_dtm)

svcacc=metrics.accuracy_score(ytest1,ypred)
svcprec=metrics.precision_score(ytest1,ypred)
svcrec=metrics.recall_score(ytest1,ypred)
print("accuracy=",svcacc)
print("Recall=",svcrec)
print("Precision=",svcprec)
print("Confusion Matrix\n",metrics.confusion_matrix(ytest1,ypred))
```

### Random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=300)
clf = clf.fit(xtr_dtm,ytrain1)
ry_pred = clf.predict(xte_dtm)
racc=metrics.accuracy_score(ytest1,ry_pred)
rprec=metrics.precision_score(ytest1,ry_pred)
rrec=metrics.recall_score(ytest1,ry_pred)
print("accuracy=",racc)
print("Recall=",rrec)
print("Precision=",rprec)
print("Confusion Matrix\n",metrics.confusion_matrix(ytest1,ry_pred))
```
